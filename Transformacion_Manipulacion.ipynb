# Proyecto Desafío 2: Administración de Condominios

```python
import pandas as pd
```

```python
condominios = pd.read_json('datos_desafio_2/dados_locacao_imoveis.json')
condominios.head()
```

```python
condominios = pd.json_normalize(condominios['dados_locacao'])
condominios
```

### Convirtiendo a datatime las columnas con fechas

```python
condominios['datas_combinadas_pagamento'] = condominios['datas_combinadas_pagamento'].apply(
    lambda lista: [pd.to_datetime(fecha, dayfirst=True).date() for fecha in lista]
)
```

```python
condominios['datas_de_pagamento'] = condominios['datas_de_pagamento'].apply(
    lambda lista: [pd.to_datetime(fecha, dayfirst=True).date() for fecha in lista]
)
```

### Calculando dias de retraso

Esta funcion: 
* recorre las listas **fechas_pagadas y fechas_acordadas** al mismo tiempo
* **(pago - acordado).days** extrae los dias de retraso

```python
# Calcular retrasos en días
def calcular_retrasos(fechas_acordadas, fechas_pagadas):
    return [(pago - acordado).days for pago, acordado in zip(fechas_pagadas, fechas_acordadas)]
```

Se crea una nueva columna llamada **dias_retraso**, donde se aplica la funcion anterior fila por fila.
* Ejemplo: 
    * Fechas acordadas para pagar **[01/06/2022, 01/07/2022]**
    * Fechas donde se pago **[05/06/2022, 03/07/2022]**
* Resultado: **[4,2]**, ya que de **01/06/2022 al 05/06/2022** se tiene 4 dias de retraso, y de **01/07/2022 a 03/07/2022** son 2 dias de retraso

```python
condominios['dias_retraso'] = condominios.apply(
    lambda row: calcular_retrasos(row['datas_combinadas_pagamento'], row['datas_de_pagamento']),
    axis=1
)
```

```python
# Indicar si hubo retraso
condominios['pago_con_retraso'] = condominios['dias_retraso'].apply(
    lambda lista: [dias > 0 for dias in lista]
)
```

```python
condominios.head()
```

```python
condominios = condominios.explode(['datas_combinadas_pagamento','datas_de_pagamento','valor_aluguel','dias_retraso', 'pago_con_retraso'])
```

```python
condominios['valor_aluguel'] = condominios['valor_aluguel'].apply(lambda x: x.replace('reais','').replace(',','.').replace('$','').strip()).astype(np.float64)
```

```python
condominios
```

#### Parte 3, elimminar (blocoAP)

```python
condominios['apartamento'] = condominios['apartamento'].str.replace('[blocoAP\(\)]','',regex=True)
```

```python
condominios
```

```python
media_atraso = condominios.groupby(['apartamento'])['dias_retraso'].mean()
```
